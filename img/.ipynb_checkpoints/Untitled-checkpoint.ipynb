O ambiente é criado através do arquivo Dockerfile (gerado automaticamento pelo sits.rep) onde é especificado o tipo de sistema operacional, bibliotecas e arquivos que irão compor este ambiente. Ao executar este Dockerfile, um ambiente virtual é criado para o usuário, permitindo a interação com este ambiente.

O pacote sits.rep cria um ambiente computacional auto contido através dos dados de contexto obtidos durante a execução dos \textit{scripts}. Por tanto, são instaladas todas as dependências de sistema operacional, pacotes, com as suas respectivas versões no momento da execução original, além de configurar a semente do gerador de números aleatórios e outras informações de sistema. Nesta máquina virtual, o \textit{script} reprodutível é executado e é criado um arquivo \textit{hash} reprodutível dos resultados. No final do processo, o arquivo \textit{hash} original é comparado com o arquivo reprodutível, para verificar se os resultados produzidos são exatamente os mesmos do experimento original.
